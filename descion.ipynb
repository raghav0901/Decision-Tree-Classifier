## Decision Tree Classifier ( from scratch)

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np
import math
data = load_iris()
xtrain,xtest,ytrain,ytest=train_test_split(data.data,data.target,random_state=1)
dff = pd.DataFrame(xtrain, columns=data.feature_names)
dff=pd.concat([dff,pd.Series(ytrain)],axis=1)
dff.rename(columns = {dff.columns[-1]:'Target'}, inplace = True)


#Function used to return the gain ratio achieved on splitting on a particular feature
def split_feature(df,f):
    Yset=set(df['Target'])
    ## calculating initital entropy:
    initial_entropy=0
    for x in Yset:
        newdf=df[df['Target']==x]
        pval=(len(newdf.index))/(len(df.index))
        initial_entropy+= (-1*pval*math.log2(pval))
    ## calculating final weighted entropy:
    final=0
    features=set(df[f])
    split_info=0
    for fr in features:
        newdf=df[df[f]==fr]
        split_val=(len(newdf.index))/(len(df.index))
        split_info+=(-1*split_val*math.log2(split_val))
        Yset=set(newdf['Target'])
        initialy=0
        for x in Yset:
          newdf1=newdf[newdf['Target']==x]
          pval=(len(newdf1.index))/(len(newdf.index))
          initialy+= (-1*pval*math.log2(pval))
        final+=(initialy)*(len(newdf.index)/len(df.index))
    ## calculating and returning the test_gain_ratio
    return ((initial_entropy-final)/(split_info))  

# Function used for the recursive call and the main call:
def spliter(df,level):
    print()
    print('Level ',level)
    Yset=set(df['Target'])
    ## Printing the value counts:
    valconts=df['Target'].value_counts()
    for x in valconts.index:
        print("Count of ",x,'=',valconts[x])
    
    ## calculating and printing initital entropy:
    initial_entropy=0
    for x in Yset:
        newdf=df[df['Target']==x]
        pval=(len(newdf.index))/(len(df.index))
        initial_entropy+= (-1*pval*math.log2(pval))
    print('Current Entropy is=',initial_entropy)
    if len(Yset)>1 and len(df.columns[:-1])>0:
        features=df.columns[:-1]
        maxgain=0
        best_feature=None
        for f in features:
            test_gain=split_feature(df,f)
            if test_gain>maxgain:
                maxgain=test_gain
                best_feature=f
        tr=set(df[best_feature])
        
        #launching recursive calls:
        print('Splitting on Feature ',best_feature,'with gain ratio',maxgain)
        for r in tr:
            df_cop=df.copy()
            new_df=df_cop[df_cop[best_feature]==r].copy()
            new_df.drop(best_feature,inplace=True,axis=1)
            spliter(new_df,level+1)
    else:
        print('Reached Leaf Node')
        
# Starting the main call
spliter(dff,0)
```
